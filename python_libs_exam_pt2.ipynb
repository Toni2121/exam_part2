```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display

print('part B')
print('4.')
data = [
    {"cust_id": 1, "timestamp": "2024-01-01", "residence": "Tel Aviv", "grade": 8},
    {"cust_id": 2, "timestamp": "2024-01-02", "residence": "Jerusalem", "grade": None},
    {"cust_id": 1, "timestamp": "2024-01-03", "residence": "Tel Aviv", "grade": 8},
    {"cust_id": 3, "timestamp": "2024-01-04", "residence": None, "grade": 7},
    {"cust_id": 2, "timestamp": "2024-01-05", "residence": "Jerusalem", "grade": None}
]

print('A.')
unique_list = []
seen_ids = set()

for item in data:
    if item["cust_id"] not in seen_ids:
        unique_list.append(item)
        seen_ids.add(item["cust_id"])

display("Unique:", unique_list)

print('B.')
filtered_data = [entry for entry in data if None not in entry.values()]
display("no None:", filtered_data)

print('C.')
grades = [entry["grade"] for entry in data if entry["grade"] is not None]
average_grade = sum(grades) / len(grades)
for entry in data:
    if entry["grade"] is None:
        entry["grade"] = average_grade
print("with avg grade:", data)

```

```python
print('part C')
print('9.')

customers_df = pd.read_csv('customers.csv')
items_df = pd.read_csv('items.csv')
order_item_df = pd.read_csv('order_item.csv')
orders_df = pd.read_csv('orders.csv')

removed_customers = customers_df.loc[customers_df.isnull().sum(axis=1) >= 3, "id"].tolist()
display("Removed customers:", removed_customers)

display('customer df', customers_df)
display('items df', items_df)
display('order item df', order_item_df)
display('orders df', orders_df)
```

```python
print('part D')
print('6.')
order_item_with_customer = order_item_df.merge(orders_df[['order_id', 'customer_id']], left_on='order_id', right_on='order_id', how='left')
merged_df = order_item_with_customer.merge(customers_df[['id', 'age']], left_on='customer_id', right_on='id', how='left')
merged_df = merged_df.dropna(subset=['age'])
plt.hist(merged_df['age'], bins=20, edgecolor='black')
plt.title('show of purchases by customers age')
plt.xlabel('Age')
plt.ylabel('Number of Purchases')
plt.show()

print('7.')
order_item_with_price = order_item_df.merge(items_df[['id', 'item_price']], left_on='item_id', right_on='id', how='left')
order_item_with_price['price_total'] = order_item_with_price['item_price'] * order_item_with_price['quantity']
order_item_with_customer = order_item_with_price.merge(orders_df[['order_id', 'customer_id']], left_on='order_id', right_on='order_id', how='left')
order_item_with_customer_and_nation = order_item_with_customer.merge(customers_df[['id', 'nationallity']], left_on='customer_id', right_on='id', how='left')
sales_by_nationality = order_item_with_customer_and_nation.groupby('nationallity')['price_total'].sum().reset_index()
display(sales_by_nationality)

```
